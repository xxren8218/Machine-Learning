### pandas滑动窗口

- 进行预测单独拿某一天的值不合理，可以用一个滑动窗口，取一个平均值，比较平稳。

```python
%matplotlib inline 
import matplotlib.pylab
import numpy as np
import pandas as pd
```

```python
df = pd.Series(np.random.randn(600), index = pd.date_range('7/1/2021', freq = 'D', periods = 600))
```

```python
df.head()
```

```python
r = df.rolling(window = 10) # 滑动窗口的大小。默认从左边开始滑动。
r
```

```python
# r.max, r.median, r.std, r.skew, r.sum, r.var
print(r.mean()) # 小于窗口长度不会计算！为NaN
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15, 5))

df.plot(style='r--')
df.rolling(window=10).mean().plot(style='b')
```

**可以看到滑动窗口的平均值更平稳一些**
